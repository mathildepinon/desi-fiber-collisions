```python
import os
import sys

import numpy as np
from matplotlib import pyplot as plt

# plotting
plt.style.use(os.path.join(os.path.abspath('../bin'), 'plot_style.mplstyle'))
plots_dir = '../../plots/'
```

```python
plt.rcParams['axes.grid'] = False
```

```python
r = 12/10.4/2
circle1 = plt.Circle((0, 0), r, clip_on=False, fill=True, facecolor='C0', alpha=0.2)
c1 = plt.Circle((0, 0), 1/2, clip_on=False, fill=False, color='C0', alpha=0.5, ls='--')
circle2 = plt.Circle((1, 0), r, clip_on=False, fill=True, facecolor='C0', alpha=0.2)
c2 = plt.Circle((1, 0), 1/2, clip_on=False, fill=False, color='C0', alpha=0.5, ls='--')
circle3 = plt.Circle((1/2, np.sqrt(3)/2), r, clip_on=False, fill=True, facecolor='C0', alpha=0.2)
c3 = plt.Circle((1/2, np.sqrt(3)/2), 1/2, clip_on=False, fill=False, color='C0', alpha=0.5, ls='--')

fig, ax = plt.subplots(figsize=(3, 3))

ax.add_patch(circle1)
ax.add_patch(circle2)
ax.add_patch(circle3)
ax.add_patch(c1)
ax.add_patch(c2)
ax.add_patch(c3)

#ax.plot([0, 1], [0, 0], color='black', lw=0.7, ls='--', alpha=0.5)
ax.plot([0, 1/2], [0, np.sqrt(3)/2], color='black', lw=0.7, ls='--', alpha=0.5)
#ax.plot([1/2, 1], [np.sqrt(3)/2, 0], color='black', lw=0.7, ls='--', alpha=0.5)

ax.hlines(np.sqrt(3)/2, xmin=1/2-r, xmax=1/2+r, color='black', lw=0.7, alpha=0.5)

#x = [0, 1, 1/2]
#y = [0, 0, np.sqrt(3)/2]
#circles = [plt.Circle((xi,yi), radius=0.107/10.4/2., linewidth=0) for xi,yi in zip(x,y)]
#from matplotlib import collections
#c = collections.PatchCollection(circles)
#ax.add_collection(c)

ax.scatter([0, 1, 1/2], [0, 0, np.sqrt(3)/2], s=7)

#ax.scatter([0.7, 1.3, 0.3, -0.2, 0.63, 0.711, -0.02], [0.2, 0.1, -0.1, 0.3, 0.65, 0.83, -0.3], s=7, color='red')
#ax.scatter([0.4, 0.2, 0.92], [0.9, 0.1, 0.06], s=7, color='green')

ax.text(0, 0.25, '10.4 mm', horizontalalignment='left', fontsize=12, rotation=60)
ax.text(1/2, np.sqrt(3)/2+0.05, '12 mm', horizontalalignment='center', fontsize=12)

ax.set_aspect('equal')
ax.axis('off')
plt.savefig(os.path.join(plots_dir, 'fiber_positioners_scheme.png'), dpi=300)
plt.show()
```
