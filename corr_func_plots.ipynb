```python
import os

import numpy as np
from matplotlib import pyplot as plt

from corr_plots import mock, plot_fiber_collisions, plot_delta_fiber_collisions

plt.style.use(os.path.join(os.path.abspath(''), 'plot_style.mplstyle'))
```

## ELG

```python
nmocks = 25
mocks = [mock(id, "NGC", 'ELG', rebin=(4, 1)) for id in range(nmocks)]
ells = (0, 2, 4)
plot_fiber_collisions(mocks, mp=ells)
plt.savefig('corr_FA_effect_NGC_ELG.png')
plt.show()

mocks = [mock(id, "SGC", 'ELG', rebin=(4, 1)) for id in range(nmocks)]
plot_fiber_collisions(mocks, mp=ells)
plt.savefig('corr_FA_effect_SGC_ELG.png')
plt.show()
```

```python
mocks = [mock(id, "NGC", 'ELG', rebin=(4, 1)) for id in range(nmocks)]
eNGC = plot_delta_fiber_collisions(mocks, ells)
fig = plt.gcf()
for ax in fig.get_axes():
    ax.set_xlim((0, 30))
plt.savefig('corr_FA-complete_NGC_ELG_lowscales.png')
plt.show()

mocks = [mock(id, "SGC", 'ELG', rebin=(4, 1)) for id in range(nmocks)]
eSGC = plot_delta_fiber_collisions(mocks, ells)
fig = plt.gcf()
for ax in fig.get_axes():
    ax.set_xlim((0, 30))
plt.savefig('corr_FA-complete_SGC_ELG_lowscales.png')
plt.show()
```
